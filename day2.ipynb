## –°–æ—Ä—Ç–∏—Ä–æ–≤–∫–∞

## –ó–∞–¥–∞–Ω–∏–µ 2.2

–¢–µ–ø–µ—Ä—å —Å–¥–µ–ª–∞–µ–º —Ç–æ –∂–µ —Å–∞–º–æ–µ, —Ç–æ–ª—å–∫–æ –∫–ª—é—á –∫ "–∫–æ—Ä–∑–∏–Ω–∫–µ" –±—É–¥–µ—Ç `customer_id`. –î—Ä—É–≥–∏–º–∏ —Å–ª–æ–≤–∞–º–∏, –º—ã —Ö–æ—Ç–∏–º –ø–æ–¥—Å—á–∏—Ç–∞—Ç—å, –∫–∞–∫–æ–π –¥–æ—Ö–æ–¥ –ø—Ä–∏–Ω–µ—Å –Ω–∞–º –∫–∞–∂–¥—ã–π –∫–ª–∏–µ–Ω—Ç. –ù–∞–ø—Ä–∏–º–µ—Ä, –ú–∞—à–∞ –∑–∞–ø–ª–∞—Ç–∏–ª–∞ 10 000 —Ä—É–±–ª–µ–π –∑–∞ –≤—Å–µ –≤—Ä–µ–º—è —Å—É—â–µ—Å—Ç–≤–æ–≤–∞–Ω–∏—è –º–∞–≥–∞–∑–∏–Ω–∞, –∞ –°–∞—à–∞ 5 000 —Ä—É–±–ª–µ–π –∏ —Ç.–¥.

## –ú–µ–Ω—å—à–∏–Ω—Å—Ç–≤–æ ¬´–¥–µ–ª–∞–µ—Ç¬ª –±–æ–ª—å—à–∏–Ω—Å—Ç–≤–æ, –∏ –Ω–∞–æ–±–æ—Ä–æ—Ç.

–Ø –æ–±–µ—â–∞–ª –ø–æ–∫–∞–∑–∞—Ç—å –¢–µ–±–µ, —á—Ç–æ –º–µ–Ω—å—à–∏–Ω—Å—Ç–≤–æ –ø–æ—Ä–æ–∂–¥–∞–µ—Ç –±–æ–ª—å—à–∏–Ω—Å—Ç–≤–æ –∏ –Ω–∞–æ–±–æ—Ä–æ—Ç. –ú—ã –æ—á–µ–Ω—å –±–ª–∏–∑–∫–∏ –∫ —ç—Ç–æ–º—É, –Ω–∞–º –ø—Ä–æ—Å—Ç–æ –Ω—É–∂–Ω–æ –ø–æ–ª—É—á–∏—Ç—å —Å—É–º–º—É –∏–Ω–¥–∏–≤–∏–¥—É–∞–ª—å–Ω—ã—Ö –¥–æ—Ö–æ–¥–æ–≤ –¥–ª—è –∫–ª–∏–µ–Ω—Ç–æ–≤ (—ç—Ç–æ –∑–∞–¥–∞–Ω–∏–µ 2.2). 

–î–∞–ª–µ–µ - –¥–æ–±–∞–≤–ª—è—è –∫–∞–∂–¥—É—é –ø–æ—Å–ª–µ–¥—É—é—â—É—é —Å—Ç—Ä–æ–∫—É (–∑–∞–∫–∞–∑) –≤–º–µ—Å—Ç–µ, –Ω–∞–ø—Ä–∏–º–µ—Ä, –ø–æ—Å–ª–µ 10 —Å—Ç—Ä–æ–∫, –º—ã –ø–æ–ª—É—á–∏–º –æ–±—â—É—é —Å—É–º–º—É –∑–∞–∫–∞–∑–æ–≤ –¥–ª—è –ø–µ—Ä–≤—ã—Ö –¥–µ—Å—è—Ç–∏ –∫–ª–∏–µ–Ω—Ç–æ–≤. –ö–æ–≥–¥–∞ –º—ã —Å–ª–æ–∂–∏–º –≤—Å–µ —Å—Ç—Ä–æ–∫–∏, –º—ã –ø–æ–ª—É—á–∏–º –≤—Å—é –≤—ã—Ä—É—á–∫—É –º–∞–≥–∞–∑–∏–Ω–∞.

–î–∞–≤–∞–π –ø–æ–∫–∞–∂—É –Ω–∞ –∫–æ–¥–µ - —Ç–∞–∫ –±—É–¥–µ—Ç –ø—Ä–æ—â–µ –æ–±—ä—è—Å–Ω–∏—Ç—å, —á—Ç–æ –∏–º–µ—é –≤ –≤–∏–¥—É.

–£ –Ω–∞—Å –µ—Å—Ç—å —Ñ—É–Ω–∫—Ü–∏—è –¥–ª—è —ç—Ç–æ–≥–æ `cumsum()`.

````python
sorted_orders = (
    df[ ["customer_id", "price_total"] ]
     .groupby("customer_id")
     .agg('sum')
     .sort_values(by="price_total", ascending=False)
    .reset_index()
)

sorted_orders["price_cumsum"] = sorted_orders["price_total"].cumsum()

#sorted_orders.head(10)
sorted_orders
````

–ü–æ—è–≤–∏–ª–∞—Å—å –Ω–æ–≤–∞—è —Å—Ç—Ä–æ—á–∫–∞ `sorted_orders["price_cumsum"] = sorted_orders["price_total"].cumsum()`

## –ó–∞–¥–∞–Ω–∏–µ 2.3

–ò—Å–ø–æ–ª—å–∑—É—è –ø—Ä–∞–≤–∏–ª–æ 80/20, –º—ã –ø—Ä–∏–º–µ–Ω–∏–ª–∏ –µ–≥–æ –¥–ª—è –∫–ª–∏–µ–Ω—Ç–æ–≤, —Ç–µ–ø–µ—Ä—å –ø–æ—Ä–∞ —Å–¥–µ–ª–∞—Ç—å —Ç–æ –∂–µ —Å–∞–º–æ–µ –¥–ª—è –ø—Ä–æ–¥—É–∫—Ç–æ–≤. –î—Ä—É–≥–∏–º–∏ —Å–ª–æ–≤–∞–º–∏, –º—ã —Ö–æ—Ç–∏–º –ø—Ä–æ–≤–µ—Ä–∏—Ç—å, —Ä–∞–±–æ—Ç–∞–µ—Ç –ª–∏ —ç—Ç–æ –ø—Ä–∞–≤–∏–ª–æ –∏ –¥–ª—è –ø—Ä–æ–¥—É–∫—Ç–æ–≤.

–í–æ–∑–Ω–∏–∫–∞–µ—Ç –≤–æ–ø—Ä–æ—Å: –ø—Ä–∏–Ω–æ—Å–∏—Ç –ª–∏ –º–µ–Ω—å—à–∏–Ω—Å—Ç–≤–æ –ø—Ä–æ–¥—É–∫—Ç–æ–≤ (–æ–∫–æ–ª–æ 20%) –±–æ–ª—å—à—É—é —á–∞—Å—Ç—å –¥–æ—Ö–æ–¥–æ–≤ (80%)?


## üí° –°–æ–≤–µ—Ç
* –í —ç—Ç–æ–º —Å–ª—É—á–∞–µ –≤–º–µ—Å—Ç–æ `customer_id`, –¥–æ–ª–∂–µ–Ω –±—ã—Ç—å –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω `product_id`.
* –•–æ—Ä–æ—à–æ –ø–æ–¥—É–º–∞–π, –∏ —É –¢–µ–±—è –≤—Å–µ –ø–æ–ª—É—á–∏—Ç—Å—è. –ó–¥–µ—Å—å, –¥–µ–π—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–æ, –µ—Å—Ç—å —á—Ç–æ "–Ω–µ–º–Ω–æ–≥–æ" –∏–∑–º–µ–Ω–∏—Ç—å. –ö–∞–∫ –≤–∞—Ä–∏–∞–Ω—Ç, —Å–¥–µ–ª–∞–π –Ω–µ–±–æ–ª—å—à–æ–π –ø–µ—Ä–µ—Ä—ã–≤, –≤–æ–∑–º–æ–∂–Ω–æ, –æ—Ç–¥–æ—Ö–Ω–∏ –∫–∞–∫-–Ω–∏–±—É–¥—å –∏–Ω–∞—á–µ üõÄ –∏–ª–∏ –±–æ–ª–µ–µ –∞–∫—Ç–∏–≤–Ω–æ üèäüèºüö¥üèá –∏ —Å–Ω–æ–≤–∞ –≤–µ—Ä–Ω–∏—Å—å –∫ –∑–∞–¥–∞—á–µ. –¢—ã –æ–±—è–∑–∞—Ç–µ–ª—å–Ω–æ —ç—Ç–æ —Å–¥–µ–ª–∞–µ—à—å!

````python
sorted_orders = (
    df[ ["product_id", "price_total"] ]
     .groupby("product_id")
     .agg('sum')
     .sort_values(by="price_total", ascending=False)
    .reset_index()
)

sorted_orders["price_cumsum"] = sorted_orders["price_total"].cumsum()


p80 = int(df["price_total"].sum() * 0.8)
top_products =sorted_orders[ sorted_orders["price_cumsum"] < p80 ]["product_id"]
len(top_products)/df["product_id"].nunique()
````
